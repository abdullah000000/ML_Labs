```python
# create a Python class with a constructor and functions of all access specifiers each.
class MyClass:
    def __init__(self, name):
        self.name = name
        print("Hello World Parent")
    def public(self):
        print("Public")
    def _protected(self):
        print("Protected")
    def __private(self):
        print("Private")
```

```python
obj = MyClass("Hello World")
```

```python
#main function to call all the functions of the class
if __name__ == '__main__':
    obj.public()
    obj._protected()
    # obj.__private() # this will give an error
    
```

```python
# Inherit a subclass from the above created class, initiate its
# constructor that also initiates the parent classâ€™ constructor. 
class MySubClass(MyClass):
    def __init__(self, name2):
        super().__init__(name2)
        self.name2 = name2

    # Add a function in this subclass
    def my_function(self):
        print("Child Function")
```

```python
# Create an object of parent class and call its functions. Create an object of child class and try accessing child functions as well as parent class functions.
obj1 = MyClass("1")
obj1.public()
obj1._protected()
# obj1.__private() # this will give an error
obj2 = MySubClass("2")
obj2.my_function()
```

```python
# Given a list of strings, return the count of the number of strings where the string length is 2 or more and the first and last chars of the string are the same
def match_words(words):
    count = 0
    for word in words:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count
```

```python
print(match_words(['def', 'abc', 'ded', '3443']))
```

Question # 04

```python
# Given a list of strings, return a list with the strings in sorted order,except group all the strings that begin with 'x' first.
# e.g. ['mix', 'xyz', 'apple', 'xanadu', 'aardvark'] yields ['xanadu', 'xyz', 'aardvark', 'apple', 'mix']
# Hint: use sorted() and lambda.
def front_x(words):
    return sorted(words, key=lambda x: x[0] == 'x') + sorted(words, key=lambda x: x[0] != 'x')
```

```python
print(front_x(['mix', 'xyz', 'apple', 'xanadu', 'aardvark']))
```

Question # 05

```python
# Given a list of non-empty tuples, return a list sorted in increasing order by the last element in each tuple.
def sort_last(tuples):
    return sorted(tuples, key=lambda x: x[-1])
```

```python
print(sort_last([(1, 7), (1, 3), (3, 4, 5), (2, 2)]))
```

Question # 06

```python
# Given a list of numbers, return a list where all adjacent == elements have been reduced to a single element, so [1, 2, 2, 3] returns [1, 2, 3].
def remove_adjacent(nums):
    return [nums[i] for i in range(len(nums)) if i == 0 or nums[i - 1] != nums[i]]
```

```python
print(remove_adjacent([1, 2, 2, 3]))
```

```python
# two lists sorted in increasing order, create and return a merged list of all the elements in sorted order
def linear_merge(list1, list2):
    return sorted(list1 + list2)
```

```python
print(linear_merge(['aa', 'xx', 'zz'], ['bb', 'cc']))
```
