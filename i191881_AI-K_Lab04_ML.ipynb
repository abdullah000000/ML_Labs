```python
# Assuming 3 friends have 5 cities (A,B,C,D,E) to go to, what is the possible state space of this problem? Display the possible state space in a proper way.
count =0
# create an array for the cities
cities = ['A', 'B', 'C', 'D', 'E']
# create an array for the friends
friends = ['1', '2', '3']
# display possible state space for 3 friends who can go to 5 cities
print('Possible state space for 3 friends who can go to 5 cities:')
for i in range(len(cities)):
    for j in range(len(cities)):
        for k in range(len(cities)):
                    print(cities[i], cities[j], cities[k])
                    count = count + 1
print('Total number of possible state space:', count)




```

```python
# find random variable where two friends go to the same city
count_P = 0
print('Random variable where two friends go to the same city:')
for i in range(len(cities)):
    for j in range(len(cities)):
        for k in range(len(cities)):
            if cities[i] == cities[j] or cities[i] == cities[k] or cities[j] == cities[k]:
                count_P += 1
                print(cities[i], cities[j], cities[k])
print(count_P)
```

```python
# find random variable where no two friends go to the same city
count_Q = 0
print('Random variable where no two friends go to the same city:')
for i in range(len(cities)):
    for j in range(len(cities)):
        for k in range(len(cities)):
            if cities[i] != cities[j] and cities[i] != cities[k] and cities[j] != cities[k]:
                count_Q += 1
                print(cities[i], cities[j], cities[k])
print(count_Q)
```

```python
# find the probability of two friends going to the same city
print('Probability of two friends going to the same city:')
print(count_P/125)
print('Probability of no two friends going to the same city:')
print(count_Q/125)
    
        
                    
```

```python
# plot probability distribution for two friends going to the same city and no two friends going to the same city
plt.bar(['2 friends in same city', 'No friend in same city'], [count_P/125, count_Q/125])
plt.show()


    
```

```python
# Pre-processing the data
# Convert categorical data to numerical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['mainroad'] = le.fit_transform(df['mainroad'])
df['guestroom'] = le.fit_transform(df['guestroom'])
df['basement'] = le.fit_transform(df['basement'])
df['hotwaterheating'] = le.fit_transform(df['hotwaterheating'])
df['airconditioning'] = le.fit_transform(df['airconditioning'])
df['prefarea'] = le.fit_transform(df['prefarea'])
df['furnishingstatus'] = le.fit_transform(df['furnishingstatus'])

```

```python
df
```

```python
# apply Random forest classifier on the given dataset
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# split the data into train and test
X = df.drop(['furnishingstatus'], axis=1)
y = df['furnishingstatus']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# create a random forest classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=0)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)

            
# print the accuracy score
print('Accuracy score:', accuracy_score(y_test, y_pred))

```
